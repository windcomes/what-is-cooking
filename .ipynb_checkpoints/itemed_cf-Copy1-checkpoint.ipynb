```python
def sim_ingredient_matrix():
    sim_ingredient = np.zeros((n_ingredient,n_ingredient))
    
    for i in range(n_ingredient):
        related_i = []
        for r in range(n_recipes):
            if recipe_ingredient[r,i]==1:
                for ii in range(i+1,n_ingredient):
                    related_i.append(ii)
        related_i = list(set(related_i))
        related_i.sort()
        for ii in related_i:
            sim_ingredient[i,ii] = sim(set_r_for_i[all_ingredient[i]],set_r_for_i[all_ingredient[ii]])
            sim_ingredient[ii,i] = sim_ingredient[i,ii]
        
        print(all_ingredient[i])
    return(sim_ingredient)

sim_ingredient = sim_ingredient_matrix()
```

```python
print(sim_ingredient)

np.set_printoptions(suppress=True)
def top_n(n):
   top_n_matrix = np.zeros((n_ingredient,n))
   for i in range(n_ingredient):
       top_n_matrix[i,:] = np.argsort((-sim_ingredient[i]))[:n]
   return top_n_matrix

top_n_matrix=top_n(10)
print(top_n_matrix)
```

```python
def print_similiar_items(partial_recipe):
    print('\nwe recommend these ingredients based on %s:' % partial_recipe)
    rank =[]
    for j in all_ingredient:
        num = 0
        if j not in partial_recipe:
            p_j = 0
            s_j = top_n_matrix[all_ingredient.index(j)]

            for i in partial_recipe:
                if  all_ingredient.index(i) in s_j:
                    p_j = p_j + sim_ingredient[all_ingredient.index(j),all_ingredient.index(i)]
                    num = num + 1
            if num>0:
                rank.append((p_j/num,j))
          
    rank.sort(reverse=True)
    for i in range(10):
          print(rank[i][1],rank[i][0])
          
partial_recipe = [
      "ampalaya",
      "salt",
      "luke warm water",
      "ground black pepper",
      "onions"]
print_similiar_items(partial_recipe)
```
